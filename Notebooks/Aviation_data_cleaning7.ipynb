```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned_aviation_data_US.csv
df = pd.read_csv('working-df/cleaned_aviation_data_US_v3.csv')

df.info()
```

```python
# Show Reason value_counts including NaN
df['Reason'].value_counts(dropna=False)
```

```python
# Create a subset of records with 'pilot error' as Reason
pilot_error_df = df[df['Reason'] == 'pilot error']
pilot_error_df.info()
```

```python
# Show me Report_Status value_counts that are not 'Probable Cause' or 'Factual'
df_report_known = df[~df['Report_Status'].isin(['Probable Cause', 'Factual'])]
```

```python
# Show df_report_known Report_Status value_counts that are not 'pilot error' under Reason
df_report_known[df_report_known['Reason'] != 'pilot error'].info()
```

```python
df_report_known[df_report_known['Reason'] != 'pilot error']['Report_Status'].value_counts()
```

```python
# replace 'pilots' with "pilot's" in Report_Status column
df['Report_Status'] = df['Report_Status'].str.replace('pilots', 'pilot\'s')
```

```python
df[df['Reason'] != 'pilot error']['Report_Status'].value_counts()
```

```python
# If Report_Status contains "pilot's", fill in Reason as 'pilot error'
df.loc[df['Report_Status'].str.contains('pilot\'s'), 'Reason'] = 'pilot error'
```

```python
df[df['Reason'] != 'pilot error']['Report_Status'].value_counts()
```

```python
# If Report_Status is 'Probable Cause', 'Unknown', or 'Factual', fill in Reason as Unknown
df.loc[df['Report_Status'].isin(['Probable Cause', 'Unknown', 'Factual']), 'Reason'] = 'Unknown'
```

```python
# Show reason values including NaN
df['Reason'].value_counts(dropna=False)
```

```python
# Show report_status values that have Reason value as NaN
df[df['Reason'].isna()]['Report_Status'].value_counts()
```

```python
# Show report_status values over 10 that have Reason value as NaN
df[df['Reason'].isna()]['Report_Status'].value_counts()[df[df['Reason'].isna()]['Report_Status'].value_counts() > 10]
```

```python
# If Report_Status is 'Preliminary', 'Foreign', fill in Reason as Unknown
df.loc[df['Report_Status'].isin(['Preliminary', 'Foreign']), 'Reason'] = 'Unknown'
```

```python
# Show report_status values over 10 that have Reason value as NaN
df[df['Reason'].isna()]['Report_Status'].value_counts()[df[df['Reason'].isna()]['Report_Status'].value_counts() > 10]
```

```python
# If Report_Status contains 'loss of engine' fill in Reason as 'Mechanical Failure'
df.loc[df['Report_Status'].str.contains('loss of engine'), 'Reason'] = 'Mechanical Failure'
```

```python
# Show report_status values that have Reason value as NaN
df[df['Reason'].isna()]['Report_Status'].value_counts()
```

```python
# For NaN Reason values, fill in as 'Unknown'
df['Reason'].fillna('Unknown', inplace=True)
```

```python
# Export df as new CSV
df.to_csv('cleaned_aviation_data_US_v4.csv', index=False)
```
